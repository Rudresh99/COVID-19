```python
import numpy as np
import pandas as pd
from datetime import datetime
import io
import requests

url1="https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
s1=requests.get(url1).content
Confirmed_frame=pd.read_csv(io.StringIO(s1.decode('utf-8')))

url2="https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"
s2=requests.get(url2).content
Death_frame=pd.read_csv(io.StringIO(s2.decode('utf-8')))

url3="https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv"
s3=requests.get(url3).content
Recovery_frame=pd.read_csv(io.StringIO(s3.decode('utf-8')))

print("Confirmed_frame Shape :",Confirmed_frame.shape)
print("Death_frame Shape : ",Death_frame.shape)
print("Recovery Frame : ",Recovery_frame.shape)
print("Done Sir")
Confirmed_frame.head()
#Death_frame.head()
#Recovery_frame.head()


```

```python
raw_data_Confirmed = pd.melt(Confirmed_frame, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], var_name=['Date'])
raw_data_Deaths = pd.melt(Death_frame, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], var_name=['Date'])
raw_data_Recovered = pd.melt(Recovery_frame, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], var_name=['Date'])

print("Confirmed data : ",raw_data_Confirmed.shape)
print("Daeth data: ",raw_data_Deaths.shape)
print("Recovered data : ",raw_data_Recovered.shape)
raw_data_Confirmed.head()
```

```python
raw_data_Confirmed['Date'] = pd.to_datetime(raw_data_Confirmed['Date'])
raw_data_Deaths['Date'] = pd.to_datetime(raw_data_Deaths['Date'])
raw_data_Recovered['Date'] = pd.to_datetime(raw_data_Recovered['Date'])
raw_data_Confirmed.head()
```

```python
raw_data_Confirmed.columns = raw_data_Confirmed.columns.str.replace('value', 'Confirmed')
raw_data_Deaths.columns = raw_data_Deaths.columns.str.replace('value', 'Deaths')
raw_data_Recovered.columns = raw_data_Recovered.columns.str.replace('value', 'Recovered')
raw_data_Deaths.head()
```

```python
raw_data_Confirmed.isnull().sum()
```

```python
raw_data_Confirmed['Province/State'].fillna(raw_data_Confirmed['Country/Region'], inplace=True)
raw_data_Deaths['Province/State'].fillna(raw_data_Confirmed['Country/Region'], inplace=True)
raw_data_Recovered['Province/State'].fillna(raw_data_Confirmed['Country/Region'], inplace=True)

raw_data_Confirmed.isnull().sum()
```

```python
print("The Shape of Cornirmed is: ", raw_data_Confirmed.shape)
print("The Shape of Cornirmed is: ", raw_data_Deaths.shape)
print("The Shape of Cornirmed is: ", raw_data_Recovered.shape)
```

```python

full_join = raw_data_Confirmed.merge(raw_data_Deaths[['Province/State','Country/Region','Date','Deaths']], 
                                      how = 'outer', 
                                      left_on = ['Province/State','Country/Region','Date'], 
                                      right_on = ['Province/State', 'Country/Region','Date'])

print("Shape of first join: ", full_join.shape)


full_join = full_join.merge(raw_data_Recovered[['Province/State','Country/Region','Date','Recovered']], 
                                      how = 'outer', 
                                      left_on = ['Province/State','Country/Region','Date'], 
                                      right_on = ['Province/State', 'Country/Region','Date'])

print("Shape of second join: ", full_join.shape)

full_join.head()
```

```python
full_join.isnull().sum()
```

```python
full_join['Month-Year'] = full_join['Date'].dt.strftime('%b-%Y')
full_join.head()
```

```python
test = full_join[full_join['Province/State'] == 'Anhui']

 
full_join2 = test.copy()


full_join2['Date - 1'] = full_join2['Date'] + pd.Timedelta(days=1)
full_join2.rename(columns={'Confirmed': 'Confirmed - 1', 'Deaths': 'Deaths - 1', 'Recovered': 'Recovered - 1',
                          'Date': 'Date Minus 1'}, inplace=True)


full_join3 = test.merge(full_join2[['Province/State', 'Country/Region','Confirmed - 1', 'Deaths - 1', 
                            'Recovered - 1', 'Date - 1', 'Date Minus 1']], how = 'outer',
                             left_on = ['Province/State','Country/Region','Date'], 
                             right_on = ['Province/State', 'Country/Region','Date - 1'])


full_join3['Confirmed Daily'] = full_join3['Confirmed'] - full_join3['Confirmed - 1']


test.head()
full_join2.head()
full_join3.head()
```
